```python
import numpy as np
from scipy.optimize import minimize

import matplotlib.pyplot as plt
%matplotlib inline
```

## Задание

1) генерация точек на прямой с шумом
2) подбере веса w, так чтобы сумма ... была минимальной 

```python
def f(x):
    return 3 * x + 4

X = np.arange(-10, 10, 0.1)
Y = [f(x) + np.random.normal(scale=3) for x in X]

def min_func(w):
    return sum([(w[0] * x + w[1] - y) ** 2 for x, y in zip(X, Y)])

w = minimize(min_func, [1,1]).x
print w
Y_res = [x * w[0] + w[1] for x in X]
```

```python
plt.plot(X, Y)
plt.plot(X, Y_res)
plt.show()
```

```python
def min_func_abs(w):
    return sum([abs(w[0] * x + w[1] - y) for x, y in zip(X, Y)])
```

```python
Y_noise = Y
for i in xrange(10, 50):
    Y_noise[i * 3] += abs(np.random.normal(scale=20))
    
w_noise = minimize(min_func, [1,1]).x
print w_noise
Y_res_noise_1 = [x * w_noise[0] + w_noise[1] for x in X]
```
